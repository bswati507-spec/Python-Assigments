### Question 19
Find common elements between two sets.

```python
# write your answer here
# Two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

common_elements = set1.intersection(set2)
print(common_elements)  
```

### Question 20

```python
# write your answer here
# Two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

unique_elements = set1.union(set2)
print(unique_elements)  

```
