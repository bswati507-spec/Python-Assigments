## 1. What Are Conditionals?
Conditionals help Python make decisions based on conditions.

## 2. Basic `if` Example

```python
age = 20
if age >= 18:
    print("Adult")
```

## 3. `if` and `else` Example

```python
age = 15
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

## 4. `elif` Example

```python
marks = 72
if marks >= 90:
    print("A Grade")
elif marks >= 75:
    print("B Grade")
else:
    print("C Grade")
```

## 5. Comparison Operators

| Operator | Meaning |
|---------|---------|
| == | equal to |
| != | not equal to |
| > | greater than |
| < | less than |
| >= | greater than or equal |
| <= | less than or equal |

## 6. Simple Examples

```python
num = 6
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
temp = 40
if temp > 35:
    print("It's hot")
else:
    print("Normal weather")
```

# Practice Questions

Given:

### bmi = 27


Use conditions:

BMI < 18.5 → "Underweight"

BMI < 25 → "Normal"

Else → "Overweight"

```python
#Answer here
bmi=27

if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normal")
else:
    print("Overweight")
```

## Q. Electricity Bill Alert


### units = 310


Print:

“Normal Usage” if units <= 200

“High Usage” if units <= 300

“Very High Usage” otherwise

```python
#answer here
units = 310
if units <= 200:
    print("Normal Usage")
elif units <= 300:
    print("High Usage")
else:
    print("Very High Usage")
```
